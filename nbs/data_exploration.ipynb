```python
import os

import numpy as np
import pandas as pd
```

```python
DATA_DIR = '/project/cq-training-1/project2/data/'

with open(os.path.join('low-resource-translation', DATA_DIR, 'train.lang1'), 'r') as f:
    train_en = f.readlines()

with open(os.path.join('low-resource-translation', DATA_DIR, 'train.lang2'), 'r') as f:
    train_fr = f.readlines()
    
print(f'len of train_en : {len(train_en)}')
print(f'len of train_fr : {len(train_fr)}')
```

```python
train_en[100], train_fr[100]
```

```python
def get_words(sentences):
    words = []
    for sentence in sentences:
        words += sentence.split(' ')
    return np.array(words)

words = get_words(train_en)
```

```python
print(f'Number of words : {len(words)}')
print(f'Number of unique words : {len(np.unique(words))}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

def plt_horizontal_bar(bar_data, labels, title='', filename='horizontal_bar'):
    plt.figure(figsize=(10, 15))
    y_pos = np.arange(len(labels))
    plt.barh(y_pos, bar_data, height=0.25, tick_label=labels)
    plt.xlim(0.0, np.max(bar_data))
    plt.title(title)
    plt.show()
#     plt.savefig(os.path.join(PLT_FOLDER,filename+'.png'), bbox_inches='tight')

unique_words, word_counts = np.unique(words, return_counts=True)

sorted_idx = np.argsort(word_counts)[::-1]
sorted_word_counts = word_counts[sorted_idx]
sorted_unique_words = unique_words[sorted_idx]

nb = 50
plt_horizontal_bar(sorted_word_counts[:nb][::-1], sorted_unique_words[:nb][::-1], title='50 tokens that appear most often in the english training data')
```

```python
import os
import numpy as np

DATA_DIR = '/project/cq-training-1/project2/data/'

with open(os.path.join(DATA_DIR, 'train.lang1'), 'r') as f:
    train_en = f.readlines()

def get_words(sentences):
    words = []
    for sentence in sentences:
        words += sentence.strip().split(' ')
    return np.array(words)

words = get_words(train_en)

print(f'Number of words : {len(words)}')
print(f'Number of unique words : {len(np.unique(words))}')
```
